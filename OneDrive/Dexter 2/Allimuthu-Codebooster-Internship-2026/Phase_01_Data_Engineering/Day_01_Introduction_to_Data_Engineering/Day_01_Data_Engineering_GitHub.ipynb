```python
import pandas as pd

df = pd.read_csv("student_performance.csv")

SEP = "=" * 60
sep = "-" * 60

print(SEP)
print("       STUDENT PERFORMANCE EXPLORER")
print("         Powered by Pandas & Python")
print(SEP)

print("\nDATASET OVERVIEW")
print(sep)

print("Total Students        :", len(df))
print("Total Columns         :", len(df.columns))
print("Unique Departments    :", df["department"].nunique())

print("\nDEPARTMENT-WISE STUDENT COUNT")
print(sep)

dept_counts = df["department"].value_counts()

for dept, count in dept_counts.items():
    print(dept, ":", count, "students")

print("\nSCORE ANALYSIS")
print(sep)

print("Highest Math Score        :", df["math_score"].max())
print("Lowest Math Score         :", df["math_score"].min())
print("Average Math Score        :", round(df["math_score"].mean(), 2))
print("Average Programming Score :", round(df["programming_score"].mean(), 2))
print("Average Attendance        :", round(df["attendance_percentage"].mean(), 2), "%")

print("\nGENDER-WISE AVERAGE MATH SCORE")
print(sep)

gender_avg = df.groupby("gender")["math_score"].mean()

for gender, avg in gender_avg.items():
    print(gender, ":", round(avg, 2))

print("\nTOP 5 STUDENTS IN MATH")
print(sep)

top5 = df.sort_values("math_score", ascending=False).head(5)

print("Name\t\tDepartment\tMath Score")

for _, row in top5.iterrows():
    print(row["name"], "\t", row["department"], "\t", row["math_score"])

print("\nLOW ATTENDANCE STUDENTS (<75%)")
print(sep)

low_att = df[df["attendance_percentage"] < 75]

if low_att.empty:
    print("All students have attendance above 75%")
else:
    print("Name\t\tDepartment\tAttendance")

    for _, row in low_att.iterrows():
        print(row["name"], "\t", row["department"], "\t", str(row["attendance_percentage"]) + "%")

print("\n" + SEP)
```
