````python
1. ML (Day 5) vs Generative AI
Traditional ML trains models on labeled data to make predictions or classifications — for example, predicting house prices or detecting spam.
It outputs a number, a class, or a probability. Generative AI, built on large language models, is trained on massive text corpora and can generate new content — writing, code, summaries, or answers — based on a prompt.
 ML solves narrowly defined tasks; Generative AI handles open-ended, language-based tasks without task-specific training.
````

````python
2. What does temperature=0.0 do?
Temperature controls randomness in the model's output. At 0.0, the model always picks the highest-probability token, making responses fully deterministic and consistent. You would use it when you need reliable, reproducible outputs — such as data extraction, JSON parsing, classification, or any task where creativity is unwanted and correctness matters most.
 Higher values (e.g. 0.8) introduce variety, useful for creative writing or brainstorming.
````

````python
4. LLM Hallucination & how prompt engineering reduces it
Hallucination is when an LLM generates text that sounds confident but is factually wrong or completely fabricated — inventing names, dates, citations, or
figures that do not exist. Prompt engineering reduces it by grounding the model: providing the actual source text to extract
from (rather than asking it to recall), using instructions like "only use information from the passage below", setting temperature=0.0 for consistency,
and asking the model to respond with "I don't know" if unsure rather than guessing.
````

````python
5. Fix for json.loads() crashing on markdown-wrapped JSON
The LLM wraps its output in a code fence (```json ... ```), which json.loads() cannot parse directly. Strip the fences first:
pythonimport json
import re

raw = "```json\n{\"name\": \"Ramesh\"}\n```"
cleaned = re.sub(r"```json\s*|\s*```", "", raw).strip()

data = json.loads(cleaned)
print(data)
re.sub removes the opening ```json and closing ``` markers, leaving only the pure JSON string for safe parsing.

````
