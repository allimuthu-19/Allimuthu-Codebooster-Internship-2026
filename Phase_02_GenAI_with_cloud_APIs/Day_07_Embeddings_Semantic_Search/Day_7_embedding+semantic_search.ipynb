```python
import pandas as pd
import numpy as np

from sentence_transformers import SentenceTransformer
import chromadb

print("All libaries imported successfully")
print(f"Chromadb version: {chromadb.__version__}")
print(f"Pandas version: {pd.__version__}")
```

```python
doucments = [
    "ETL is used to clean and transform data",
    "A vehicle is a mode of transportation",
    "Cars and trucks are popular automobiles",
    "SQL is used to query databases",
    "Machine learning trains models on data"
]

query_keyword = "vehicle"
print(f"KEYWORD SEARCH for:'{query_keyword}'")
print("="*60)

for i , doc in enumerate(doucments):
  if query_keyword.lower() in doc.lower():
    print(f"FOUND [ doc_{i}]:{doc}")
  else:
    print(f"NOT FOUND [ doc_{i}]:{doc}")
print()
print("PROBLEM: doc_2 talks about 'Cars and trucks'- which ARE vehicles! ")
print("but keyword search MISSED it because it searched for the exact word 'vehicle'")
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model loaded successfully")
print(f"Model produces vectors of sizes:{model.get_sentence_embedding_dimension()}")
```

```python
sentence= "ETL is used to clean and transform data"
embedding = model.encode(sentence)
print(f"Input sentence:'{sentence}")
print()
print(f"Embedding type:{type(embedding)}")
print(f"Embedding shape:{embedding.shape}")
print(f"First 10 numbers:{embedding[:10]}")
print(f"Min value:{embedding.min():.4f}")
print(f"Man value:{embedding.max():.4f}")
```

```python
sentence= [ "ETL is used to clean and transform data",
    "A vehicle is a mode of transportation",
    "Cars and trucks are popular automobiles",
    "SQL is used to query databases",
    "Machine learning trains models on data"
]
embedding = model.encode(sentence)
print(f"Input sentence:{sentence}")
print(f"number of sentences:{len(sentence)}")
print(f"Shape of embediings array:{embedding.shape}")
print()
```

```python
def cosine_similarity(vec_a, vec_b):
  dot_product = np.dot(vec_a, vec_b)

  norm_a = np.linalg.norm(vec_a)
  norm_b = np.linalg.norm(vec_b)

  return dot_product / (norm_a * norm_b)

sim_01 = cosine_similarity(embedding[0],embedding[1])
sim_02 = cosine_similarity(embedding[0],embedding[2])
sim_12 = cosine_similarity(embedding[1],embedding[2])
print("="*60)
print(f"sentence 0: '{sentence[0]}'")
print(f"sentence 1: '{sentence[1]}'")
print(f"sentence 2: '{sentence[2]}'")
print()
print(f"Similarity (0 vs 1): {sim_01:.4f}")
print(f"Similarity (0 vs 2): {sim_02:.4f}")
print(f"Similarity (1 vs 2): {sim_12:.4f}")
print()
print("their cosine similarity is score is high")
```
