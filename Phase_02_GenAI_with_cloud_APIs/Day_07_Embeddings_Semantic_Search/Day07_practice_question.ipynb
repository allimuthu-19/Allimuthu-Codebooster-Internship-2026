```python
Q3. Model A to add, Model B to query вЂ” will results be correct?
No, results will be incorrect (or meaningless).
Each model maps text into its own unique vector space. Model A places "cat" at vector [0.2, 0.8, ...] while Model B places it at [0.9, -0.3, ...] вЂ” completely different coordinates.
 Comparing them via cosine/dot-product similarity produces random, unreliable rankings.
 You must use the same model consistently for both indexing and querying.
```
