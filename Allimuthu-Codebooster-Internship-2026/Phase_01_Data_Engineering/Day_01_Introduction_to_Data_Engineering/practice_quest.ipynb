```python
import pandas as pd
df = pd.read_csv('student_performance.csv')
```

### Students from Electronics department

```python
e_students = df[df['department'] == 'Electronics']
print(e_students[['name', 'department']])
```

### Average attendance of female students

```python
female = df[df['gender'] == 'Female']
print(f"The average percentage of attendance of female students is {female['attendance_percentage'].mean():.2f}%")
```

### student with lowest programming score

```python
print(f"Student with lowest programming score is {df.loc[df['programming_score'].idxmin(), 'name']} with a score of {df['programming_score'].min()}")
```

### students scored above 85 in mathematics

```python

high_math_students = df[df['math_score'] > 85]
print("Students who scored above 85 in mathematics:")
print(high_math_students[['name', 'math_score']])
```

### average total_score for students with attendance above 90%

```python
df['total_score'] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']
attendance_above_90 = df[df['attendance_percentage']>90]
print(attendance_above_90['total_score'].mean())
```
